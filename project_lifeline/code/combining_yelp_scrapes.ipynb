```python
import pandas as pd
import phonenumbers
```

```python
lena_scrape = pd.read_csv('../data/yelp_first_pass_cleaning_zip_lena.csv')
paul_scrape = pd.read_csv('../data/yelp_first_pass_cleaning_zip_paul.csv')
sj_scrape = pd.read_csv('../data/yelp_first_pass_cleaning_zip_sj.csv')
boston_scrape = pd.read_csv('../data/yelp_first_pass_cleaning.csv')
suffolk = pd.read_csv('../data/yelp_suffolk_186_first_pass_cleaning_zip.csv')
```

```python
yelp_scrape = pd.concat([lena_scrape, paul_scrape, sj_scrape, boston_scrape, suffolk], axis = 0, ignore_index = True)
yelp_scrape.head()
```

```python
print(yelp_scrape.shape)
print(lena_scrape.shape)
print(paul_scrape.shape)
print(sj_scrape.shape)
print(boston_scrape.shape)
print(suffolk.shape)
```

```python
yelp_scrape.drop_duplicates(['id'], keep="first", inplace = True)
yelp_scrape.shape
```

```python
yelp_scrape.head()
```

```python
yelp_scrape.tail()
```

```python
yelp_scrape.phone = yelp_scrape.phone.astype(str).str.strip('0').str.strip('.')
```

```python
for i in yelp_scrape.index:
    try:
        yelp_scrape.loc[i, 'phone']= phonenumbers.format_number(phonenumbers.parse(yelp_scrape.loc[i, 'phone'], 'US'), 'US')
    except:
        pass
```

```python
yelp_scrape.head()
```

```python
yelp_scrape.to_csv('../data/final_yelp_scrape.csv', index = False)
```
