```python
import pandas as pd
import numpy as np
import time
import requests
```

```python
url = " https://api.yelp.com/v3/businesses/search"

#you need to enter your own API KEY
api_key = "enter your api key here"

headers = {'Authorization': 'bearer %s' % api_key}
```

```python
categories = {'communications':[["banks", 'radiostations', 'televisionstations']],
              'energy':[['fueldocks', 'servicestations', 'evchargingstations', "propane"]],
             'food_water_shelter': [["pickyourown", "ranches", "waterpurification", 'rvparks', 'resorts'],
                                   ["bedbreakfast", "guesthouses", "healthretreats", "hostels", "hotels"]],
             "hazmat": [["biohazardcleanup", "enviroabatement", "environmentaltesting", "hazardouswastedisposal"]],
             "health_medical": [["funeralservices", "emergencypethospital", "vets", "medicalsupplies", "acupuncture"],
                               ["animalassistedtherapy", "assistedliving", "behavioranalysts", "blooddonation", "bodycontouring"],
                               ["cannabis_clinics", "chiropractors", "colonics", "conciergemedicine", "c_and_mh"],
                               ["crisispregnancycenters"], ["cryotherapy"], ["dentalhygienists"], ["dentists"], ["diagnosticservices"],
                               ["dialysisclinics", "doctors", "doulas", "emergencyrooms", "habilitativeservices"],
                               ["healthcoach", "homehealthcare", "hospice", "hospitals", "ivhydration"],
                               ["lactationservices", "liceservices", "medcenters", "medicalfoot", "medicaltransportation"],
                               ["midwives", "nursepractitioner", "occupationaltherapy", "optometrists", "organdonorservices"],
                               ["orthotics", "personalcare", "pharmacy", "physicaltherapy", "postpartumcare"],
                               ["prenatal", "prosthetics", "prosthodontists", "rehabilitation_center", "reproductivehealthservices"],
                               ["retirement_homes", "skillednursing", "ultrasoundimagingcenters"]],
             "safe_sec": [["collegeuniv", "elementaryschools", "highschools", "montessori", "preschools"],
                         ["privateschools", "religiousschools", "specialed", "waldorfschools", "artschools"],
                         ["nursingschools", "publicservicesgovt"]],
             "transportation": [["airports", "carrental", "motorcycle_rental", "rvrental", "trainstations"], 
                               ["transport", "bikerentals", "marinas", "busrentalA10	"]]}
```

```python
final_yelp_df = pd.DataFrame()

for category in categories.keys():
    for i in range(len(categories[category])):
        offset = 0
        total_results_old = 0
        while True:
            parameter = {'categories': categories[category][i], 'limit': 50, 'radius': 30000, 'location':'Boston',
                         'offset': offset}
            res = requests.get(url = url, headers = headers, params = parameter)
            time.sleep(10)
            if res.status_code != 200:
                print('error when requestion {}'.format(categories[category][i]))
                break
            results = res.json()
            results_df = pd.DataFrame(results['businesses'])
            results_df['lifeline'] = category
            final_yelp_df = pd.concat([final_yelp_df, results_df], axis = 0, ignore_index = True, sort = True)
            total_results = len(final_yelp_df.name.unique())
            if total_results_old == total_results:
                break
            else:
                total_results_old = total_results
                offset += 50
        
final_yelp_df
```

```python
terms = {'food_water_shelter':[['food'], ['restaurants']]}

terms_df = pd.DataFrame()

for term in terms.keys():
    for i in range(len(terms[term])):
        offset = 0
        total_results_old = 0
        while True:
            parameter = {'term': terms[term][i], 'limit': 50, 'radius': 30000, 'location':'Boston',
                         'offset': offset}
            res = requests.get(url = url, headers = headers, params = parameter)
            time.sleep(10)
            if res.status_code != 200:
                print('error when requestion {}'.format(terms[term][i]))
                break
            results = res.json()
            results_df = pd.DataFrame(results['businesses'])
            results_df['lifeline'] = term
            terms_df = pd.concat([terms_df, results_df], axis = 0, ignore_index = True, sort = True)
            total_results = len(final_yelp_df.name.unique())
            if total_results_old == total_results:
                break
            else:
                total_results_old = total_results
                offset += 50
        
terms_df
```

```python
final_yelp_df = pd.concat([final_yelp_df, terms_df], axis = 0, ignore_index = True, sort = True)
```

```python
#saving the product information in to a csv for further cleaning and analysis
final_yelp_df.to_csv('./data/yelp_raw_info.csv', index = False)
```

```python
yelp_clean = pd.DataFrame()
```

```python
#dropping some columns that are not useful in our case
yelp_clean[['lifeline', 'name', 'url', 'phone', 'distance', 'price', 'id']] = final_yelp_df[['lifeline',
                                                                                       'name', 'url', 'phone',
                                                                          'distance', 'price', 'id']]
```

```python
#unpacking location
yelp_clean['zipcode'] = [final_yelp_df.location[i]['zip_code'] for i in range(len(final_yelp_df.location))]
yelp_clean['city'] = [final_yelp_df.location[i]['city'] for i in range(len(final_yelp_df.location))]
yelp_clean['state'] = [final_yelp_df.location[i]['state'] for i in range(len(final_yelp_df.location))]
yelp_clean['address1'] = [final_yelp_df.location[i]['address1'] for i in range(len(final_yelp_df.location))]
yelp_clean['address2'] = [final_yelp_df.location[i]['address2'] for i in range(len(final_yelp_df.location))]
yelp_clean['address3'] = [final_yelp_df.location[i]['address3'] for i in range(len(final_yelp_df.location))]
yelp_clean['display_address'] = [final_yelp_df.location[i]['display_address'][0] for i in range(len(final_yelp_df.location))]
#unpacking coordinates
yelp_clean['latitude'] = [final_yelp_df.coordinates[i]['latitude'] for i in range(len(final_yelp_df.location))]
yelp_clean['longitude'] = [final_yelp_df.coordinates[i]['longitude'] for i in range(len(final_yelp_df.location))]
yelp_clean['alias1'] = [final_yelp_df.categories[i][0]['alias'] for i in range(len(final_yelp_df.location))]
try:
    yelp_clean['alias2'] = [final_yelp_df.categories[i][1]['alias'] for i in range(len(final_yelp_df.location))]
except:
    yelp_clean['alias2'] = 'NA'
try:
    yelp_clean['alias3'] = [final_yelp_df.categories[i][2]['alias'] for i in range(len(final_yelp_df.location))]
except:
    yelp_clean['alias3'] = 'NA'
```
